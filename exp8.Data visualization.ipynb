```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

Line chart

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

Bar chart

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
plt.bar(x,y, color='orange')
```

scatter plot

```python
a=(1,3,4,5,6,8,9,2,6,7)
b=(10,25,30,45,68,95,46,25,14,35)
plt.scatter(a,b)
plt.title("scatter chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```
